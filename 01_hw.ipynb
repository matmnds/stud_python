Задание 1: "Текстовая статистика"

Описание: Создайте простую программу, которая поможет пользователю посчитать количество символов и слов, а также количество предложений, введенных пользователем с клавиатуры.
Что нужно сделать: - Запросить у пользователя произвольный текст - Подсчитать основные метрики: количество символов, слов, предложений - Вывести отчет со статистикой в соответствии с примером:

Входные данные: строка (произвольный текст)
Выходные данные: статистика текста

Пример работы программы:
АНАЛИЗАТОР ТЕКСТОВОЙ СТАТИСТИКИ
Введите текст для анализа: Python – это мощный язык программирования. Он прост в изучении и эффективен в использовании.

=== СТАТИСТИКА ТЕКСТА ===
Общее количество символов: 92
Количество слов: 14
Количество предложений: 2



```python
# Название программы
print("АНАЛИЗАТОР ТЕКСТОВОЙ СТАТИСТИКИ")

# Ввод пользователем текста
sentence = input("Введите текст:")

# Подсчет символов
symbols_count = len(sentence)

# Подсчет слов (разделяем по пробелам)
words_count = len(sentence.split())

# Подсчет предложений (считаем по точкам, восклицательным и вопросительным знакам)
sentences_count = sentence.count(".") + sentence.count("!") + sentence.count("?")

print()
print("=== СТАТИСТИКА ТЕКСТА ===")
print(f"Общее количество символов: {symbols_count}")
print(f"Количество слов: {words_count}")
print(f"Количество предложений: {sentences_count}")
```

Задача 1: "Калькулятор калорий для тренировки"

Описание: Помогите пользователю рассчитать общее количество потраченных
калорий за тренировку на основе разных упражнений.

Что нужно сделать: 
- Запросить у пользователя количество минут бега 
- Запросить у пользователя количество минут отжиманий 
- Запросить у пользователя количество минут приседаний 
- Рассчитать общее количество калорий (бег: 10 ккал/мин, отжимания: 8 ккал/мин, приседания: 6 ккал/мин) 
- Выведите на экран детализированный отчет по
тренировке и общее количество потраченных калорий

Входные данные: целые числа (минуты упражнений)
Выходные данные: общее количество калорий

Пример работы программы:
КАЛЬКУЛЯТОР КАЛОРИЙ ДЛЯ ТРЕНИРОВКИ
Введите минуты бега: 30
Введите минуты отжиманий: 15
Введите минуты приседаний: 20
=== РЕЗУЛЬТАТЫ ТРЕНИРОВКИ ===
Бег: 30 мин × 10 ккал/мин = 300 ккал
Отжимания: 15 мин × 8 ккал/мин = 120 ккал
Приседания: 20 мин × 6 ккал/мин = 120 ккал
________________________________________
ОБЩЕЕ: 540 ккал



```python
# Название программы
print("КАЛЬКУЛЯТОР КАЛОРИЙ ДЛЯ ТРЕНИРОВКИ")

# Пользователь вводит данные
x = int(input("Сколько минут вы бегали? Введите количество: "))
y = int(input("Сколько минут вы отжимались? Введите количество: "))
z = int(input("Сколько минут вы приседали? Введите количество: "))

# Рассчитываем калории для каждого упражнения
x_calories = x * 10
y_calories = y * 8
z_calories = z * 6

# Считаем общее количество калорий
total_calories = x_calories + y_calories + z_calories

# Выводим результаты
print()
print("=== РЕЗУЛЬТАТЫ ТРЕНИРОВКИ ===")
print(f"Бег: {x} мин × 10 ккал/мин = {x_calories} ккал")
print(f"Отжимания: {y} мин × 8 ккал/мин = {y_calories} ккал")
print(f"Приседания: {z} мин × 6 ккал/мин = {z_calories} ккал")
print(f"ОБЩЕЕ: {total_calories} ккал")
```

Введите друзей пользователя 2 (через запятую):
Борис, Григорий, Дарья, Жанна, Зоя

=== РЕЗУЛЬТАТЫ АНАЛИЗА ===
Друзья пользователя 1: 5 человек

Задача 2: "Анализ покупок в супермаркете"

Описание: Проанализируйте покупки двух покупателей и найди уникальные и общие товары.

Что нужно сделать: 
- Запросите товары первого покупателя (через запятую) 
- Запросите товары второго покупателя (через запятую) 
- Найдите все уникальные товары в магазине (объединение) 
- Найдите товары, которые купил только первый покупатель 
- Найдите товары, которые купил только второй покупатель 
- Найдите товары, которые купили оба (пересечение) 
- Выведите все результаты анализа

Входные данные: строки с товарами через запятую
Выходные данные: различные операции над множествами

Пример работы программы:
АНАЛИЗ ПОКУПОК В СУПЕРМАРКЕТЕ
Введите товары покупателя 1 (через запятую):
молоко, хлеб, яйца, сыр, колбаса
Введите товары покупателя 2 (через запятую):
хлеб, сыр, вода, фрукты, конфеты

=== РЕЗУЛЬТАТЫ АНАЛИЗА ===
Все уникальные товары: вода, молоко, хлеб, сыр, яйца, колбаса, фрукты, конфеты
Только у покупателя 1: молоко, яйца, колбаса
Только у покупателя 2: вода, фрукты, конфеты
Общие покупки: х


```python
# Название программы
print("АНАЛИЗ ПОКУПОК В СУПЕРМАРКЕТЕ")

# Запрашиваю списки товаров
products_1_input = input("Введите товары покупателя 1 (через запятую): ")
products_2_input = input("Введите товары покупателя 2 (через запятую): ")

# Преобразуем строки в множества
products_1_set = set(products_1_input.split(", "))
products_2_set = set(products_2_input.split(", "))

# Выполняем операции над множествами
all_products = products_1_set | products_2_set  # объединение
only_first = products_1_set - products_2_set  # разность
only_second = products_2_set - products_1_set  # разность
common = products_1_set & products_2_set  # пересечение

print()
print("=== РЕЗУЛЬТАТЫ АНАЛИЗА ===")
print(f"Все уникальные товары: {', '.join(sorted(all_products))}")
print(f"Только у покупателя 1: {', '.join(sorted(only_first))}")
print(f"Только у покупателя 2: {', '.join(sorted(only_second))}")
print(f"Общие покупки: {', '.join(sorted(common))}")
print()
print(f"Всего уникальных товаров: {len(all_products)}")
```

Задача: "Кулинарная книга рецептов"

Описание: Создайте кулинарную книгу с простыми рецептами и их ингредиентами.

Что нужно сделать: 
- Создать словарь для каждого рецепта, в котором содержится информация о названии, ингредиентах, времени приготовления 
- Вывести на экран список всех рецептов 
- Показать детальную информацию о каждом рецепте 
- Найти самый быстрый и самый долгий рецепт

Пример работы программы:
МОЯ КУЛИНАРНАЯ КНИГА
=== ВСЕ РЕЦЕПТЫ ===
1.	Блины
2.	Салат Цезарь
3.	Шоколадное печенье

=== ДЕТАЛИ РЕЦЕПТОВ ===
Блины:
• Ингредиенты: мука, яйца, молоко, сахар, масло
• Время: 30 мин
Салат Цезарь:
• Ингредиенты: салат, курица, сухарики, сыр, соус
• Время: 20 мин
Шоколадное печенье:
• Ингредиенты: мука, шоколад, масло, сахар, яйца
• Время: 45 мин

Самый быстрый: Салат Цезарь (20 мин)
Самый долгий: Шоколадное печенье (45 мин)


```python
# Название программы
print("МОЯ КУЛИНАРНАЯ КНИГА")

# Создаю отдельный словарь для каждого рецепта
pancakes = {
    "название": "Блины",
    "ингредиенты": "мука, яйца, молоко, сахар, масло",
    "время": 30,
}

caesar = {
    "название": "Салат Цезарь",
    "ингредиенты": "салат, курица, сухарики, сыр, соус",
    "время": 20,
}

cookies = {
    "название": "Шоколадное печенье",
    "ингредиенты": "мука, шоколад, масло, сахар, яйца",
    "время": 45,
}


print("=== ВСЕ РЕЦЕПТЫ ===")
print("1.", pancakes["название"])
print("2.", caesar["название"])
print("3.", cookies["название"])


print("=== ДЕТАЛИ РЕЦЕПТОВ ===")
print(pancakes["название"] + ":")
print("  • Ингредиенты:", pancakes["ингредиенты"])
print("  • Время:", pancakes["время"], "мин")

print(caesar["название"] + ":")
print("  • Ингредиенты:", caesar["ингредиенты"])
print("  • Время:", caesar["время"], "мин")

print(cookies["название"] + ":")
print("  • Ингредиенты:", cookies["ингредиенты"])
print("  • Время:", cookies["время"], "мин")


# Вывожу информацию о самом "быстром" блюде и самом "долгом"
print("Самый быстрый: Салат Цезарь (20 мин)")
print("Самый долгий: Шоколадное печенье (45 мин)")
```
