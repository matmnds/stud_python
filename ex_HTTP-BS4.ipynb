# Анализ товаров интернет-магазина через API"

## Цель задания:

Практическое знакомство с основами работы с HTTP-запросами в Python, парсингом JSON-данных и анализом полученной информации.

## Условие задачи:
Вы — стажёр в отделе аналитики. Ваша первая задача: подключиться к тестовому API интернет-магазина, получить список товаров и провести по ним базовый анализ.

Вам предоставили доступ к API по адресу: https://dummyjson.com/products

## Что нужно сделать:
* Получите данные: Используя библиотеку requests, отправьте GET-запрос по указанному URL.
* Проверьте ответ: Убедитесь, что запрос выполнен успешно (код ответа 200).
* Извлеките информацию: Преобразуйте ответ в удобный формат (словарь Python) и извлеките следующие данные:
    * Общее количество товаров (total).
    * Список самих товаров (products).
* Проанализируйте и выведите результат: Напишите код, который:
    * Выводит общее количество товаров в каталоге.
    * Выводит название (title) и цену (price) для первых 15 товаров.

(Дополнительное задание) Найти и вывести самый дорогого товара в полученной выборке и его цену.

```python
!pip install requests
import requests

# Отправляю get запрос
url = "https://dummyjson.com/products"
response = requests.get(url)

# Проверяю ответ - должен быть 200
if response.status_code != 200:
    print("Ошибка")
    exit()

# Преобразую данные в словарь
data = response.json()

# Извлекаю данные по общему кол-ву товаров и списку товаров
total_products = data["total"]
products = data["products"]

# Вывожу сводную информацию:
print(f"Общее количество товаров: {total_products}")
print("Первые 15 товаров:")


# Вывожу первые 15 товаров
for i, product in enumerate(products[:15], 1):
    title = product["title"]
    price = product["price"]
    print(f"{i:2}. {title:50} - ${price}")

# Доп. задание - вывести самый дорогой товар и его цену
if products:
    most_expensive = max(products, key=lambda x: x["price"])
    print(f"Самый дорогой товар: {most_expensive['title']}")
    print(f"Цена: ${most_expensive['price']}")
```

# Парсинг интернет-магазина книг - расширенный анализ

## Цель задания:

Научиться извлекать различные типы данных с
 веб-страницы, работать с атрибутами тегов и структурировать полученную информацию.

## Условие задачи:

Вам дан начальный код для парсинга сайта книжного магазина http://books.toscrape.com/. Код уже умеет парсить названия книг с HTML-страницы.

Ваша задача — дополнить код, чтобы он также извлекал цены и рейтинги книг, и выводил всю информацию в структурированном виде.

## Что нужно сделать:

* **Изучите структуру данных**: Откройте сайт в браузере и посмотрите, как оформлены цены и рейтинги книг (через "Инспектор" в инструментах разработчика).
* **Дополните код**: Добавьте функционал для извлечения:
    * Цены книги (находится в теге с классом price_color)
    * Рейтинга книги (определяется по классам в теге <p> с классом star-rating)
* Структурируйте вывод: Измените финальный вывод программы так, чтобы для каждой книги отображалось:
    * Порядковый номер
    * Название книги
    * Цена
    * Рейтинг (в текстовом или числовом формате)

```python
# Скорректированный вариант кода

"""
ПАРСИНГ ИНТЕРНЕТ-МАГАЗИНА КНИГ
Сайт: http://books.toscrape.com/ (специально создан для обучения парсингу)
"""

# ШАГ 0: ПОДКЛЮЧАЕМ БИБЛИОТЕКИ
# ====================================================
from bs4 import BeautifulSoup 
import requests

# ШАГ 1: ПОЛУЧАЕМ HTML-КОД СТРАНИЦЫ
# ====================================================

url = "http://books.toscrape.com/"  
response = requests.get(url)        # Отправляю запрос 

# Проверяю, успешно ли выполнен запрос
if response.status_code == 200:
    print("Страница успешно загружена!")
else:
    print(f"Ошибка! Код: {response.status_code}")
    exit()

html_content = response.text  # Сохраняю HTML-код страницы в переменную


print(f"Размер загруженной страницы: {len(html_content)} символов")
print()

# ШАГ 2: СОЗДАЕМ ОБЪЕКТ ДЛЯ ЧТЕНИЯ HTML
# ====================================================

soup = BeautifulSoup(html_content, 'html.parser')

# ПОЛУЧАЕМ ЗАГОЛОВОК СТРАНИЦЫ
# -------------------------------------
print("\n1. ЗАГОЛОВОК СТРАНИЦЫ:")
page_title = soup.title.string  # Беру текст из тега <title>

print(f"Тег <title>: \n{soup.title}")      # Показывается весь тег
print(f"Текст заголовка: '{page_title}'") # Показывается только текст

# НАХОДИМ ВСЕ ЗАГОЛОВКИ h3 (ПОД ЭТИМ ТЕГОМ НАХОДЯТСЯ НАЗВАНИЯ КНИГ)
# -------------------------------------------------
print("\n2. НАЗВАНИЯ КНИГ (все теги <h3>):")
book_titles = soup.find_all('h3')  # Ищу ВСЕ теги h3 на странице

# РАЗНИЦА МЕЖДУ find() И find_all():
# find() - находит ПЕРВЫЙ подходящий элемент (или None)
# find_all() - находит ВСЕ подходящие элементы (список)

print(f"Найдено тегов h3: {len(book_titles)}")
print("Содержимое каждого h3:")

for i, title_tag in enumerate(book_titles, 1):   # Внутри каждого h3 есть тег <a> с названием книги
    book_name = title_tag.a.string  # Извлекаю текст из тега <a>
    print(f"{i}. '{book_name}'")

# СТРУКТУРА HTML ДЛЯ ПОНИМАНИЯ:
# <h3>
#   <a href="catalogue/book-title_123/index.html" title="Book Title">
#     Book Title   ← вот этот текст мы извлекаем!
#   </a>
# </h3>

# ШАГ 3: ДОПОЛНЯЕМ КОД - ИЗВЛЕКАЕМ ЦЕНЫ И РЕЙТИНГИ
# ====================================================
print("\n" + "="*70)
print("3. ПОЛНАЯ ИНФОРМАЦИЯ О КНИГАХ (первые 15):")

# Нахожу все книги - каждая в теге <article class="product_pod">
books = soup.find_all('article', class_='product_pod')

# Создаю словарь для преобразования рейтинга из текста в число
rating_translation = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

# Обрабатываю первые 15 книг
for i, book in enumerate(books[:15], 1):
    title = book.find('h3').find('a')['title']     # Извлекаю название книги (из тега h3 > a)
    price_tag = book.find('p', class_='price_color') # Извлекаю цену (из тега с классом price_color)
    price = price_tag.text if price_tag else 'Нет цены'
    rating_tag = book.find('p', class_='star-rating') # Извлекаю рейтинг (из тега с классом star-rating)
    rating_text = 'Нет рейтинга'
    
    if rating_tag:
        classes = rating_tag.get('class', [])   # Получаю все классы тега
        if len(classes) > 1:   # Второй класс содержит текстовое значение рейтинга
            rating_class = classes[1]  
            rating_text = rating_translation.get(rating_class, rating_class) # Преобразую текст в число или оставляем текстом
    
    # Вывожу информацию о книге
    print(f"{i:2}. {title}")
    print(f"     Цена: {price}")
    print(f"     Рейтинг: {rating_text}")
```
