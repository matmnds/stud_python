## Задание: Создание класса Array (как в Pascal)
Описание: Создать класс `Array`, реализовывать методы `append`, `get`, `length` и `print` как в **Pascal**.
При решении данной задачи можно описать все функции, использовать специальные (магические) методы и наследование. Ход решения - на Ваше усмотрение.

## Примерный ход решения:
* Создать класс с именем `Array`
* Создать конструктор (__init__), который должен принимать переменное количество аргументов и сохранять их в приватный атрибут (например, `_data`)
* Описать все методы `append`, `get`, `length` и `print`.
* Протестировать все созданны методы в коде 

`Ваша главная задача - научиться работать с классами, поэтому включайте воображение и экспериментируйте!Можете решить разными способами: описать все функции с нуля, использовать специальные (магические) методы и наследование от списков Python.`

```python
# Создаю класс array
class Array:

    def __init__(self, *args):  # создаю конструктор

        self._data = list(args)  # преобразую аргументы в список

    def append(self, item):  # добавляет элементы в массив

        self._data.append(item)

    def get(self, index):  # возвращает элемент по указанному индексу

        return self._data[index]

    def length(self):

        return len(self._data)

    def print(self):  # выводит содержимое массива

        print(self._data)

    def __str__(self):  # магический метод для строкового представления объекта

        return f"Array({self._data})"

    def __len__(self):  # магический метод для поддержки функции len()

        return self.length()


# Тестирую класс


def test_array_class():  # создаю функцию для тестирования всех методов класса Array

    print("ТЕСТИРОВАНИЕ КЛАССА ARRAY")

    # создаю массив с начальными значениями
    arr = Array(1, 2, 3, 4, 5)
    print(f"Создан массив: {arr}")
    print(f"Длина массива: {arr.length()}")

    # добавляю элементы в массив
    arr.append(6)
    arr.append(7)
    print(f"После добавления 6 и 7: {arr}")
    print(f"Новая длина: {arr.length()}")

    # получаю элементы по индексу
    print(f"arr.get(0) = {arr.get(0)}")  # первый элемент
    print(f"arr.get(3) = {arr.get(3)}")  # четвертый элемент
    print(f"arr.get(-1) = {arr.get(-1)}")  # последний элемент

    # применяю магические методы
    print(f"print(arr) выводит: ", end="")
    arr.print()

    print(f"str(arr) = {str(arr)}")
    print(f"len(arr) = {len(arr)}")

    # создаю пустой массив
    empty_arr = Array()
    print(f"Пустой массив: {empty_arr}")
    print(f"Длина пустого массива: {empty_arr.length()}")

    # добавляю элементы в пустой массив
    empty_arr.append(10)
    empty_arr.append(20)
    print(f"После добавления элементов: {empty_arr}")

    # тест всех методов
    print("Итоговый тест всех методов:")
    test_arr = Array("a", "b", "c")
    print(f"Исходный массив: {test_arr}")
    print(f"Длина: {test_arr.length()}")

    test_arr.append("d")
    print(f"После append('d'): {test_arr}")

    print(f"Элемент с индексом 2: {test_arr.get(2)}")

    print("Вывод с помощью метода print(): ", end="")
    test_arr.print()


# запускаю тестирование
if __name__ == "__main__":
    test_array_class()
```
