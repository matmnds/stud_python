Задача 1: «Сумма чисел»
Описание: Написать рекурсивную функцию, которая вычисляет сумму чисел от 1 до N.

Что нужно сделать:
- Создать функцию recursive_sum() с параметром n
- Реализовать базовый случай (когда n = 1)
- Реализовать рекурсивный случай (n + сумма от 1 до n-1)
- Протестировать функцию на разных значениях

```python
# Создаю рекурсивную функцию с параметром n:

def recursive_sum(n):
    if n == 1:         # базовый случай: если n = 1, возвращает 1
        return 1       
    else:
        return n + recursive_sum(n - 1) # рекурсивный случай

# Тестирую функцию на разных значениях:
try:
    n = int(input("Введите число больше или равное 1"))
    
    # Проверяю корректность ввода перед вызовом функции:
    if n < 1:
        print("Ошибка! Введите число больше или равное 1 ")
    else:
        result = recursive_sum(n)
        print(f"Сумма чисел от 1 до {n} = {result}")
except:
    print("Ошибка! Введите целое число.")
        
```

Задача 2: «Палиндром-детектор»

Описание: Напишите рекурсивную функцию, которая проверяет, 
является ли строка палиндромом (читается одинаково слева направо и справа налево). 
Игнорируйте регистр букв и пробелы.

Что нужно сделать:
- Создать рекурсивную функцию is_palindrome(text)
- Функция должна игнорировать регистр и пробелы
- Использовать базовый случай для пустой строки или строки из одного символа
- Вывести результат проверки для нескольких тестовых строк

Входные данные: строка

Выходные данные: логическое значение (True/False)

ПАЛИНДРОМ-ДЕТЕКТОР

=== ТЕСТИРОВАНИЕ ФУНКЦИИ ===
"радар" - True
"Python" - False
"A posa yпала на лапу Азора" - True
"мадам" - True
"программирование" - False

```python
# Создаю функцию: 

def is_palindrome(text):
    lower_text = text.replace(" ", "").lower()   # Привожу к нижнему регистру и убираю пробелы
    if len(lower_text) <= 1:                     # Базовый случай 1: если пустая строка или один символ
        return True
    if lower_text[0] != lower_text[-1]:          # Базовый случай 2: если первый и последний символы разные
        return False
    return is_palindrome(lower_text[1:-1])       # Рекурсивный случай: проверяю середину строки


# Тестирую функцию на разных словах:
words = ["Мадам", "довод", "роТор", "заказ", "к азак",
              "привет", "мир", "python", "программа"]

for word in words:
    result = is_palindrome(word)
    print(f'"{word}" - {result}')
```

Задача 1: «Декоратор-приветствие»
Описание: Создать простой декоратор, который добавляет приветственное сообщение перед выполнением функции.

Что нужно сделать:
- Создать декоратор greeting_decorator
- Декоратор должен выводить "Функция запускается..." перед вызовом функции
- Выводить "Функция завершена!" после выполнения функции
- Применить к двум простым функциям (на Ваше усмотрение)

Входные данные: любые функции

Выходные данные: приветственные сообщения

```python
# Создаю декоратор:
def greeting_decorator(function):
    def wrap():
        print("Функция запускается...")
        function()  
        print("Функция завершена!")
    
    return wrap


# Создаю две функции для теста: 

@greeting_decorator  # Применяем декоратор
def can_count():
    print("Привет, я умею считать сумму чисел.")

# Вторая функция
@greeting_decorator  # Применяем декоратор
def calculate_sum():
    a = 5
    b = 10
    result = a + b
    print(f"Результат: {result}")

can_count()
calculate_sum()

```

Задача 2: «Декоратор для проверки положительных чисел»
Описание: Создать декоратор, который проверяет, что все аргументы функции являются положительными числами.

Что нужно сделать:
- Создать декоратор positive_numbers_only
- Декоратор должен проверять все позиционные аргументы
- Если найден неположительный аргумент (≤ 0 ≤ 0) - выводить сообщение об ошибке
- Если все аргументы положительные - выполнять функцию

Входные данные: функции с числовыми аргументами
Входные данные: результат или сообщение об ошибке 

```python
# Создаю декоратор для проверки положительных чисел
def positive_numbers_only(func):
    def wrap(*args):                # (*args) - чтобы функция могла принять несколько аргументов
        for arg in args:
            if not isinstance(arg, (int, float)) or arg <= 0:   # проверяю, является ли аргумент положительным числом
                print(f"Ошибка: Аргумент {arg} не является положительным числом!")
                return None  # Прерываю выполнение функции 
        
        # Если все аргументы положительные, функция выполняется
        result = func(*args)
        print(f"Результат: {result}")
        return result
    
    return wrap


# Проверяю на тестовых функциях: 

# Создаю 2 функции:
@positive_numbers_only
def multiply(a, b):
    return a * b


@positive_numbers_only
def add_three(a, b, c):
    return a + b + c


multiply(3, 5)

multiply(3, -5)

add_three(2, 8, 14)

add_three(2, 8, -10)

multiply(5, 0)
```

Задача: «Генератор чётных чисел»
Описание: Создать генератор, который выдает четные числа до заданного предела.

Что нужно сделать:
- Создать функцию-генератор even_numbers()
- Использовать ключевое слово yield
- Генерировать только четные числа
- Остановиться при достижении предела

```python
# Создаю функцию-генератор четных чисел:

def even_numbers(limit):
    for number in range(2, limit + 1, 2):
        yield number


def user_input():
    try:
        limit = int(input("Введите число:"))      # Запрашиваю конечное число у пользователя
        print()
        
        if limit < 2:
            print("Ошибка! Число должно быть не меньше 2.")
            print(f"Первое четное число - 2, вы ввели: {limit}")
        else:
            print(f"Чётные числа до {limit}:")
            
            
            for even_num in even_numbers(limit):       # Генератор для вывода чисел
                print(even_num)
    except ValueError:
        print("Ошибка! Введите целое число.")

user_input()
```
