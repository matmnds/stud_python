## 1. Задание: Подсчет слов в тексте
Описание: Напишите программу, которая подсчитывает количество слов в текстовом файле. Программа должна вывести общее количество слов и количество уникальных слов (**слова с длиной < 3 символов - не учитываются**).

````python
# Открываю файл для чтения
with open('file_1.txt', 'r') as file:
    text = file.read()
    
# Разбиваю текст на слова
words = text.split()

# Фильтрую по условиям
filtered_words = [word for word in words if len(word) >= 3]

# Подсчитываю общее кол-во слов
total_words = len(filtered_words)

# Подсчитываю количество уникальных слов
unique_words = len(set(word.lower() for word in filtered_words))

print(f"Общее количество слов: {total_words}")
print(f"Количество уникальных слов: {unique_words}")
````

## 2. Задание: Фильтрация логов

Описание: У вас есть файл с логами сервера. Напишите программу, которая извлекает и сохраняет в отдельный файл строки, содержащие слово "ERROR".

*Содержимое файла (logs.txt):*
```
INFO: User logged in
ERROR: Failed to connect to database
INFO: Data processed successfully
ERROR: Timeout while fetching data
INFO: User logged out
```
*Ожидаемый результат (errors.txt):*
```
ERROR: Failed to connect to database
ERROR: Timeout while fetching data
```

````python
# Открываю файл с логами
with open('logs.txt', 'r') as log_file:
    lines = log_file.readlines()

# создаю пустой список для строк с error
error_lines = []

# Ищю строки с error
for line in lines:
    if "ERROR" in line:
        error_lines.append(line)

# Записываю найденные строки в файл
with open('errors.txt', 'w') as error_file:
    for error_line in error_lines:
        error_file.write(error_line)

# Вывожу статистику
print(f"Найдено строк с ERROR: {len(error_lines)}")
print(f"Файл errors.txt успешно создан с {len(error_lines)} строками")
````

## 3. Задание: Объединение текстов
Описание: У вас есть несколько текстовых файлов. Напишите программу, которая объединяет их содержимое в один файл. Каждый текст должен быть отделен от другого строкой с символами ---.

Содержимое файлов:

*file1.txt:*

```
The quick brown fox
```
*file2.txt:*
```
jumps over the lazy dog
```
*file3.txt:*
```
and runs away swiftly.
```
*Ожидаемый результат (combined.txt):*
```
The quick brown fox
---
jumps over the lazy dog
---
and runs away swiftly.
```

````python
# Создаю список файлов для объединения
files = ['file1.txt', 'file2.txt', 'file3.txt']

# Открываю файл для записи результата
combined_file = open('combined.txt', 'w')

# Открываю текущий файл
for index in range(len(files)):
    my_file = open(files[index], 'r')
    
    text = my_file.read()                # Читаю и записываю содержимое
    combined_file.write(text)
    
    my_file.close()                      # Закрываю текущий файл
    
    if index != len(files) - 1:          # Добавляю разделитель между тексами из разных файлов
        combined_file.write('\n---\n')

# Закрываю итоговый файл
combined_file.close()

print("Файлы успешно объединены в combined.txt")
````

## 4. Задание: Анализ данных о продажах
Описание: У вас есть CSV-файл с данными о продажах. Напишите программу, которая подсчитывает общую сумму продаж по каждому товару и выводит результат в новый CSV-файл.

*Содержимое файла (sales.csv):*
```
product,amount
apple,5
banana,3
apple,7
orange,2
banana,4
```
*Ожидаемый результат (summary.csv):*
```
product,total_amount
apple,12
banana,7
orange,2
```

````python
# Создаю словарь 
sales_summary = {}

# Открываю файл с продажами для чтения
with open('sales.csv', 'r') as sales_file:
    # Читаю заголовок и пропускаю его 
    header = sales_file.readline()
    
    # Читаю остальные строки файла
    for line in sales_file:
        parts = line.strip().split(',')               # Разделяю строку по запятой
        product = parts[0]        # название товара
        amount = int(parts[1])    # количество - преобразую в число
        
        # Добавляю количество в словарь
        if product in sales_summary:
            sales_summary[product] += amount
        else:
            sales_summary[product] = amount

# Создаю файл с итогами продаж
with open('summary.csv', 'w') as summary_file:
    summary_file.write('product,total_amount\n')        # Записываю заголовок
    for product in sorted(sales_summary.keys()):        # Записываю данные по каждому товару
        total = sales_summary[product]
        summary_file.write(f'{product},{total}\n')

# Вывожу результат на экран
print("Product, total_amount")
print("-" * 30)
for product in sorted(sales_summary.keys()):
    print(f"{product}: {sales_summary[product]}")
````
