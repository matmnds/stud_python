Задача 1: "Определитель времени суток"

Описание: Напишите программу, которая определяет время суток по введенному часу и дает рекомендацию по времяпрепровождению.

Что нужно сделать: 
- Попросить пользователя ввести текущий час (0-23) 
- Определить время суток: утро, день, вечер, ночь 
- Дать соответствующую рекомендацию

Пример работы программы:
ОПРЕДЕЛИТЕЛЬ ВРЕМЕНИ СУТОК
Введите текущий час (0-23): 14
Сейчас день!
Отличное время для работы и учебы!


```python
# Название программы
print("ОПРЕДЕЛИТЕЛЬ ВРЕМЕНИ СУТОК")


hour = int(input("Введите текущий час (0-23): "))

if hour >= 6 and hour < 12:
    print("Сейчас утро!")
    print("Самое время позавтракать и начать новый день!")
elif hour >= 12 and hour < 18:
    print("Сейчас день!")
    print("Отличное время для работы и учебы!")
elif hour >= 18 and hour < 24:
    print("Сейчас вечер!")
    print("Можно отдохнуть после тяжелого дня!")
else:
    print("Сейчас ночь!")
    print("Лучшее время для сна!")
```

Задача 1: "Таблица умножения"

Описание: Напишите программу, которая выводит на экран таблицу умножения для введённого числа с клавиатуры в красивом формате.

Что нужно сделать: 
- Попросить пользователя ввести число 
- Использовать цикл for для генерации таблицы умножения 
- Вывести результаты в формате "число × множитель = результат"

Пример работы программы:
ТАБЛИЦА УМНОЖЕНИЯ
Введите число для таблицы умножения: 7
Таблица умножения для числа 7:
7 × 1 = 7
7 × 2 = 14
7 × 3 = 21
7 × 4 = 28
7 × 5 = 35
7 × 6 = 42
7 × 7 = 49
7 × 8 = 56
7 × 9 = 63
7 × 10 = 70


```python
# Название программы 
print("ТАБЛИЦА УМНОЖЕНИЯ")

num = int(input("Введите число: "))

print(f"Таблица умножения для числа {num}:")

for i in range(1, 11):
    result = num * i
    print(f"{num} × {i} = {result}")
```

Задача 2: "Угадай число"

Описание: Программа загадывает случайное число, а пользователь пытается его угадать.

Что нужно сделать: 
- Загадать случайным образом число от 1 до 100 
- Использовать цикл while для повторения попыток 
- Давать подсказки "больше" или "меньше" - Считать количество попыток

Пример работы программы:
УГАДАЙ ЧИСЛО
Я загадал число от 1 до 100. Попробуй угадать!
Попытка 1: 50
Слишком много! Попробуй меньше.
Попытка 2: 25
Слишком мало! Попробуй больше.
Попытка 3: 37
Слишком много! Попробуй меньше.
Попытка 4: 31
Поздравляю! Ты угадал число 31 за 4 попытки!


```python
# ипортирую модуль random
import random

# Название программы 
print("УГАДАЙ ЧИСЛО")
print("Я загадал число от 1 до 100. Попробуй угадать!")

# Загадываем случайное число
any_number = random.randint(1, 100)
attempts = 0

# Цикл продолжается, пока число не угадано
while True:
    guess = int(input("Попытка: "))
    attempts += 1
    
    if guess > any_number:
        print("Слишком много! Попробуй меньше.")
    elif guess < any_number:
        print("Слишком мало! Попробуй больше.")
    else:
        print(f"Поздравляю! Ты угадал число {any_number} за {attempts} попытки!")
        break
```

Оператор case


 Задача: "Преобразователь операций с текстом"

Описание: Программа обрабатывает различные команды для работы с текстом с помощью оператора case.

Что нужно сделать: 
- Попросить пользователя ввести команду и текст 
- Использовать оператор case для обработки разных команд 

Доступные команды: 
- uppercase: преобразовать в верхний регистр 
- lowercase: преобразовать в нижний регистр 
- length: посчитать длину текста 
- words: посчитать количество слов 
- reverse: перевернуть текст 
- Выполнить соответствующую операцию с текстом 
- Вывести результат

Пример работы программы:

УМНЫЙ КАЛЬКУЛЯТОР ОПЕРАЦИЙ С ТЕКСТОМ
Доступные команды:
•	uppercase: преобразовать в верхний регистр
•	lowercase: преобразовать в нижний регистр
•	length: посчитать длину текста
•	words: посчитать количество слов
•	reverse: перевернуть текст

Введите команду: length
Введите текст: Привет, как дела?
Результат: Длина текста: 15 символов
Дополнительный пример работы:
Введите команду: words
Введите текст: Python это очень интересный язык
Результат: Количество слов: 5
Введите команду: reverse
Введите текст: hello
Результат: olleh
Введите команду: uppercase
Введите текст: привет мир
Результат: ПРИВЕТ МИР

```python
# Название программы
print("УМНЫЙ КАЛЬКУЛЯТОР ОПЕРАЦИЙ С ТЕКСТОМ")

print("Доступные команды:")
print(" - uppercase: преобразовать в верхний регистр")
print(" - lowercase: преобразовать в нижний регистр")
print(" - length: посчитать длину текста")
print(" - words: посчитать количество слов")
print(" - reverse: перевернуть текст")


command = input("Введите команду: ")
text = input("Введите текст: ")

match command:
    case "uppercase":
        print(f"Результат: {text.upper()}")
    
    case "lowercase":
        print(f"Результат: {text.lower()}")
    
    case "length":
        print(f"Результат: Длина текста: {len(text)} символов")
    
    case "words":
        print(f"Результат: Количество слов: {len(text.split())}")
    
    case "reverse":
        print(f"Результат: {text[::-1]}")
```
